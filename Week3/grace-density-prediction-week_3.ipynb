## Grace satellites

The Gravity Recovery and Climate Experiment (GRACE) twin satellites, which orbited Earth from 2002 to 2017, made detailed measurements of Earth's gravity field and improved investigations about Earth's water reservoirs, over land, ice and oceans.

GRACE measured gravity by relating it to the distance between the two satellites. When there was an increase in gravity ahead of the pair, the front satellite sped up and the distance between the pair increased. When the increased gravity was between the pair, their distance decreased; the opposite occurred when there was decreased gravity ahead of, or between, the satellite pair.

The satellites were separated by 220 km, and they could detect changes smaller than a micrometer per second in relative velocity. These measurements, in conjunction with other data and models, provided observations of terrestrial water storage changes, ice-mass variations, ocean bottom pressure changes and sea-level variations.
(https://sealevel.nasa.gov/missions/grace)



### Key Details:
1. **Mission Duration**:
   - GRACE: 2002–2017
   - GRACE-FO (Follow-On): Launched in 2018 and is currently operational.

2. **Functionality**:
   - The satellites measure variations in Earth's gravity field by monitoring the distance between them. As they pass over areas with different mass distributions, such as mountains, water reservoirs, or depleted aquifers, the gravitational pull changes, altering their relative distance.
   - The inter-satellite distance is measured using a highly precise microwave ranging system (and a laser ranging interferometer in GRACE-FO).

3. **Key Achievements**:
   - **Ice Mass Loss**: Quantified the melting of ice sheets in Greenland and Antarctica, contributing to sea-level rise studies.
   - **Groundwater Monitoring**: Detected changes in groundwater levels, helping assess water management and drought impacts.
   - **Oceanography**: Improved models of ocean currents and sea-level variations.
   - **Earthquakes and Tectonics**: Measured changes in Earth's crust due to major earthquakes.

4. **GRACE-FO Enhancements**:
   - Continuation of GRACE's mission with improved technology, including the addition of the laser ranging interferometer for higher precision.

The GRACE missions have been critical in climate science, hydrology, and geophysics, providing data that supports sustainable resource management and helps predict and adapt to environmental changes.

## Neutral Density


The GRACE satellites were not originally designed to measure neutral density directly, but they provided valuable data that could be used to infer it. Neutral density refers to the density of neutral particles in the thermosphere, a region of Earth's upper atmosphere. Understanding neutral density is essential for studying atmospheric drag on satellites and the coupling between the thermosphere and other Earth systems.

### How GRACE Contributes to Neutral Density Measurements

1. **Atmospheric Drag and Satellite Accelerometers**:
   - GRACE satellites were equipped with ultra-sensitive accelerometers that measured non-gravitational forces acting on the satellites, including atmospheric drag.
   - By analyzing the drag experienced by the satellites as they moved through the thermosphere, scientists could infer the local neutral density.

2. **Data Processing**:
   - The measured drag depends on several factors, including the satellite's velocity, the atmospheric density, and the drag coefficient.
   - Using the accelerometer data and precise orbital information, researchers estimated the neutral density along the satellite's orbit.

3. **Applications**:
   - **Thermosphere Modeling**: Improved understanding of the thermosphere's structure and dynamics, particularly its response to solar and geomagnetic activity.
   - **Space Weather Studies**: Helped quantify the impact of solar storms and geomagnetic disturbances on the upper atmosphere.
   - **Satellite Operations**: Provided critical data for predicting and mitigating drag-related orbital decay.

4. **Temporal and Spatial Coverage**:
   - GRACE's near-polar orbit and continuous data collection provided global coverage of thermospheric density variations over time, making it a valuable dataset for long-term studies.

### Limitations
- **Altitude Range**: GRACE operated at altitudes typically between 300–500 km, which limited its measurements to the lower thermosphere.
- **Indirect Measurement**: Neutral density was not measured directly but inferred from drag forces, introducing uncertainties due to assumptions about the drag coefficient and other parameters.

These combined datasets have significantly advanced our understanding of Earth's thermosphere and its coupling to solar and geomagnetic activity.

```python
# Define the score
from sklearn.metrics import mean_squared_error
import numpy as np
def print_metrics(y, y_pred):
    # 1. RMSE
    rmse = np.sqrt(mean_squared_error(np.log(y), np.log(y_pred)))
    print(f"RMSE: {rmse:.2f}")

    # 2. R
    R = np.corrcoef(np.log(y),np.log(y_pred))[0,1]
    print(f"Corr coeff: {R:.2f}")

    score = 100*np.min((R/rmse/0.74,1))
   
    print(f"Score =  {score:.2f}")

    return score
```

```python
import os
import pandas as pd
import numpy as np
from joblib import Parallel, delayed
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel, ConstantKernel

NAN_MAP ={
    'Bx' : 999.9,
    'By' : 999.9,
    'Bz' : 999.9,
    'proton_density' : 999.9,
    'temperature' : 9999999.0,
    'V': 9999.0,
    'E': 999.99,
    'beta': 999.99,
    'Dst': 99999,
}

# 4) The final score is equivalent to your grade for this assignment. It is computed by running the function print_metrics(pred,test), where "pred" is your prediction (on the test set) density, in physical units, and "test" the corresponding ground truth.
def sub_task4(model_info, x_test, y_test):
    x_test = x_test.reindex(columns=model_info['feature_names'])

    x_test_scaled = model_info['scaler'].transform(x_test)
    probs = model_info['gmm'].predict_proba(x_test_scaled)

    available = sorted(model_info['experts'].keys())
    if len(available) == 0:
        raise ValueError('No GP experts were trained in sub_task3.')

    y_log_pred = np.zeros(x_test_scaled.shape[0])
    weight_sum = np.zeros(x_test_scaled.shape[0])

    for c in available:
        w = probs[:, c]
        y_log_pred += w * model_info['experts'][c].predict(x_test_scaled)
        weight_sum += w

    weight_sum = np.clip(weight_sum, 1e-12, None)
    y_log_pred = y_log_pred / weight_sum

    y_pred = np.exp(y_log_pred)
    y_pred = np.clip(y_pred, 1e-30, None)

    score = print_metrics(y_test, y_pred)
    return y_pred, score

# 3) The dataset is too large for running GP directly on it. Split the dataset into a sufficient number of subsets and train different model independently. It is suggested that you use GMM to cluster the training data and learn the best splitting strategy.
def sub_task3(x_train, y_train, max_cluster=8, min_samples_per_cluster=1000, max_gp_samples=1000, n_jobs=None):
    y_train = np.asarray(y_train).ravel()
    if len(x_train) != len(y_train):
        raise ValueError("x_train and y_train must have the same number of samples")

    if n_jobs is None:
        n_jobs = os.cpu_count() or 1
    n_jobs = max(1, int(n_jobs))

    scaler = StandardScaler()
    x_scaled = scaler.fit_transform(x_train)

    n_samples = x_scaled.shape[0]
    if n_samples < 2:
        raise ValueError("Not enough samples to form clusters")

    k_max = min(max_cluster, n_samples, max(2, n_samples // min_samples_per_cluster))

    random_state = 2003
    ks = list(range(2, k_max + 1))

    def fit_gmm_for_k(k):
        gmm = GaussianMixture(
            n_components=k,
            covariance_type='diag',
            reg_covar=1e-6,
            random_state=random_state
        )
        gmm.fit(x_scaled)
        score = gmm.bic(x_scaled)
        return score, gmm

    gmm_results = Parallel(n_jobs=min(n_jobs, len(ks)), prefer='threads')(
        delayed(fit_gmm_for_k)(k) for k in ks
    )
    best_score, best_gmm = min(gmm_results, key=lambda t: t[0])

    pred_labels = best_gmm.predict(x_scaled)

    def train_gp_for_cluster(c):
        idx = np.where(pred_labels == c)[0]
        if idx.size == 0:
            return c, None

        fit_idx = idx
        if idx.size > max_gp_samples:
            rng = np.random.default_rng(random_state + c)
            fit_idx = rng.choice(idx, size=max_gp_samples, replace=False)

        x_subset = x_scaled[fit_idx]
        y_subset = y_train[fit_idx]

        kernel = (
            ConstantKernel(1.0, (1e-3, 1e3))
            * RBF(length_scale=np.ones(x_subset.shape[1]), length_scale_bounds=(1e-2, 1e2))
            + WhiteKernel(noise_level=1e-3, noise_level_bounds=(1e-8, 1e+1))
        )

        gp = GaussianProcessRegressor(
            kernel=kernel,
            alpha=1e-6,
            normalize_y=True,
            random_state=random_state
        )
        gp.fit(x_subset, y_subset)
        return c, gp

    expert_results = Parallel(n_jobs=min(n_jobs, best_gmm.n_components), prefer='threads')(
        delayed(train_gp_for_cluster)(c) for c in range(best_gmm.n_components)
    )
    experts = {c: gp for c, gp in expert_results if gp is not None}

    if len(experts) == 0:
        raise ValueError('No GP experts were trained. Try lowering min_samples_per_cluster or max_gp_samples constraints.')

    return {
        'experts': experts,
        'gmm': best_gmm,
        'scaler': scaler,
        'feature_names': list(x_train.columns),
    }

# 2) Hint: the target to be predicted is the column 'density'. However, since atmospheric density varies several order of magnitudes, it is recommended to transform it using its log.
def sub_task2(train_df, test_df):
    x_train = train_df.drop(columns=['density', 'time'], errors='ignore')
    y_train = np.log(train_df['density'].values)
    x_test = test_df.drop(columns=['density', 'time'], errors='ignore')
    y_test = test_df['density'].values

    return x_train, y_train, x_test, y_test

# 1) Clean the dataset: "fill values" in OMNI data are equivalent to NaNs. Remove rows containing any of those. Also, GRACE satellite data contains a flag "validity_flag". Remove all rows for which "validity_flag=1", Finally, remove any data that looks unphysical.
def sub_task1(train_df, nan_map=NAN_MAP):
    df = train_df.copy()

    if 'validity_flag' in df.columns:
        df = df[df['validity_flag'] == 0]
        df = df.drop(columns=['validity_flag'], errors='ignore')

    df = df.replace(nan_map, np.nan)

    valid = (
        np.isfinite(df['density']) & (df['density'] > 0) &
        df['latitude'].between(-90, 90) &
        df['longitude'].between(-180, 180) &
        df['local_solar_time'].between(0, 24, inclusive='left') &
        (df['proton_density'] > 0) &
        (df['temperature'] > 0) &
        (df['V'] > 0) &
        (df['beta'] >= 0)
    )

    df = df[valid]

    omni_cols = [c for c in nan_map if c in df.columns]
    df = df.dropna(subset=omni_cols).reset_index(drop=True)

    return df

def load_data():
    try:
        root_dir = os.getcwd()
        train_df = pd.read_csv(os.path.join(root_dir, 'data/grace_sat1_hourly_train.csv'))
        test_df = pd.read_csv(os.path.join(root_dir, 'data/grace_sat1_hourly_test.csv'))
    except Exception as e:
        print(f'Error loading data: {e}')
        return None, None

    return train_df, test_df

def main():
    train_data, test_data = load_data()
    print('loaded data')
    train_data = sub_task1(train_data)
    print('sub_task1 done')
    x_train, y_train, x_test, y_test = sub_task2(train_data, test_data)
    print('sub_task2 done')
    model_info = sub_task3(x_train, y_train)
    print('sub_task3 done')
    y_pred, score = sub_task4(model_info, x_test, y_test)
    print('sub_task4 done')

    return y_pred, score

main()

```
