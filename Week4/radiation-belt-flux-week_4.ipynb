## Earth's radiation belt

Radiation belt physics is a branch of space physics that focuses on understanding the dynamics of charged particles trapped by a planet's magnetic field. Earth's radiation belts, known as the **Van Allen belts**, are among the most studied examples. Below is a primer on the key concepts, structures, and processes:

---

### **Structure of the Radiation Belts**
- **Inner Radiation Belt:**
  - Located at altitudes of ~1,000–12,000 km.
  - Dominated by high-energy protons (10s of MeV) and some electrons.
  - Originates from cosmic ray interactions with Earth's atmosphere (cosmic ray albedo neutron decay, CRAND).

- **Outer Radiation Belt:**
  - Extends from ~13,000–60,000 km.
  - Dominated by high-energy electrons (up to several MeV).
  - Highly dynamic, influenced by solar activity and geomagnetic storms.

- **Slot Region:**
  - A relatively empty region between the belts (~2,000–6,000 km).
  - Cleared of particles by interactions with Earth's atmosphere and radio wave emissions (e.g., plasmaspheric hiss).

---

### **Particle Sources and Acceleration**
- **Sources of Particles:**
  - Solar wind: Electrons and protons are injected into the magnetosphere.
  - Cosmic rays: Generate secondary particles in Earth's atmosphere.
  - Magnetospheric injections: Particles from the magnetotail are energized and transported inward.

- **Acceleration Mechanisms:**
  - **Radial Diffusion:** Particles drift across magnetic field lines due to electric and magnetic field fluctuations.
  - **Wave-Particle Interactions:** Resonant interactions with electromagnetic waves (e.g., whistler-mode waves) energize particles.
  - **Substorms:** Rapid energy release in the magnetotail during geomagnetic storms injects and accelerates particles.

---

### **Loss Mechanisms**
- **Atmospheric Losses:**
  - Particles collide with atmospheric molecules, leading to energy dissipation and atmospheric precipitation.
- **Wave-Induced Scattering:**
  - Interactions with electromagnetic waves (e.g., chorus, hiss) scatter particles into the loss cone, directing them into the atmosphere.
- **Magnetopause Shadowing:**
  - Particles are lost to interplanetary space when the outer boundary of the magnetosphere shifts inward.

---

### **Key Physical Processes**
- **Adiabatic Invariants:**
  - Governs particle motion in the magnetic field:
    1. **Magnetic moment** (\( \mu \)): Conservation of energy in the gyration of particles.
    2. **Bounce motion**: Trapped particles oscillate between magnetic mirror points.
    3. **Drift motion**: Particles drift around Earth due to magnetic field gradients and curvature.

- **Wave-Particle Interactions:**
  - Whistler-mode waves: Scatter electrons into the atmosphere or accelerate them.
  - Electromagnetic ion cyclotron (EMIC) waves: Cause proton and electron loss.

---

### **Variability and Coupling**
- The belts are highly dynamic, with fluxes varying over minutes to years due to:
  - Solar wind conditions (e.g., coronal mass ejections, high-speed streams).
  - Magnetospheric substorms and geomagnetic storms.
  - Interactions with the plasmasphere and ionosphere.

---

### **Relevance and Applications**
- **Space Weather:** Understanding radiation belt dynamics is crucial for predicting space weather impacts on satellites, astronauts, and communication systems.
- **Satellite Design:** Knowledge of particle fluxes informs radiation shielding and operational planning.
- **Planetary Comparisons:** Radiation belts are observed around other planets (e.g., Jupiter, Saturn), providing comparative insights.


## Van Allen Probes

The **Radiation Belt Storm Probes (RBSP)**, later renamed the **Van Allen Probes**, were a pair of NASA spacecraft launched on **August 30, 2012**, as part of NASA's Living With a Star program. Their mission was to study the dynamics of Earth's radiation belts and improve our understanding of the processes governing the space environment. Below is an overview of the mission and its significance:

---

### **Mission Objectives**
The primary goals of the Van Allen Probes were to:
- Investigate the physical processes that accelerate and transport charged particles within the radiation belts.
- Understand the mechanisms of particle loss from the belts into Earth's atmosphere or interplanetary space.
- Examine how the belts respond to solar activity and geomagnetic storms.

---

### **Spacecraft Design**
- **Twin Spacecraft:** Two identical satellites (RBSP-A and RBSP-B) were used to provide simultaneous measurements at different locations in the radiation belts.
- **Elliptical Orbits:** The probes operated in elliptical, near-equatorial orbits ranging from ~600 km to ~32,000 km, covering both the inner and outer radiation belts.
- **Radiation-Hardened Design:** The spacecraft were designed to survive the harsh radiation environment of the belts, enabling long-term observations.

---

### **Instrumentation**
The Van Allen Probes carried a suite of instruments to measure particles, fields, and waves:
1. **Energetic Particle, Composition, and Thermal Plasma Suite (ECT):**
   - Measured the energy, composition, and distribution of charged particles.
2. **Electric and Magnetic Field Instrument Suite and Integrated Science (EMFISIS):**
   - Studied electric and magnetic fields, including wave phenomena.
3. **Radiation Belt Storm Probes Ion Composition Experiment (RBSPICE):**
   - Measured the ring current and ion composition.
4. **Relativistic Proton Spectrometer (RPS):**
   - Measured high-energy protons.
5. **Electric Field and Waves (EFW):**
   - Monitored electric fields and plasma waves.

---

### **Key Discoveries**
The Van Allen Probes made numerous groundbreaking contributions, including:
- **Discovery of a Temporary Third Radiation Belt:**
  - A transient third belt was observed during a solar storm, challenging existing models.
- **Wave-Particle Interactions:**
  - Detailed studies showed how electromagnetic waves, such as whistler-mode chorus waves, accelerate electrons to relativistic speeds.
- **Rapid Belt Changes:**
  - Observations revealed how quickly the belts can form, change, and dissipate during geomagnetic storms.
- **Loss Mechanisms:**
  - Identified mechanisms like plasmaspheric hiss and EMIC waves that scatter particles into Earth's atmosphere.

---

### **Legacy and Impact**
- The Van Allen Probes operated until **October 18, 2019**, when they were decommissioned. Their data continues to be a cornerstone for radiation belt research.
- Their findings have improved space weather forecasting and satellite design, enhancing our ability to mitigate the effects of space radiation on technology and human exploration.

```python
# Load the dataset
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import make_classification, make_blobs
%matplotlib inline

path = "./data/POES_E2_100keV_electrons_assignment_4.csv"
e2 = pd.read_csv(path)
#e2.info()

path = "./data/POES_E3_300keV_electrons_assignment_4.csv"
e3 = pd.read_csv(path)
#e3.info()

path = "./data/POES_P6_1000keV_electrons_assignment_4.csv"
p6 = pd.read_csv(path)
#p6.info()

path = "./data/OMNI_assignment_4.csv"
omni = pd.read_csv(path)
#omni.info()
    
path = "./data/RBSP_1MeV_electrons_assignment_4.csv"
rbsp = pd.read_csv(path)
#rbsp.info()

TARGET_COLS = ["RBSP_4.0","RBSP_4.1","RBSP_4.2"]

#Prepare dataset for training and testing the model
def prepare_dataset(train_df, test_df, feature_cols=None, target_cols=TARGET_COLS):
    if feature_cols is None:
        feature_cols = [col for col in train_df.columns if col not in ["Time", *target_cols]]

    x_train = train_df[feature_cols].values
    y_train = train_df[target_cols].values.astype(np.float32)

    x_test = test_df[feature_cols].values
    y_test = test_df[target_cols].values.astype(np.float32)

    scaler = StandardScaler()
    x_train = scaler.fit_transform(x_train)
    x_test = scaler.transform(x_test)

    return x_train, y_train, x_test, y_test

# Train a neural network model
def train_model(x_train, y_train, early_stopping=False, verbose=0):
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(x_train.shape[1],)),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(y_train.shape[1])
    ])

    model.compile(optimizer="adam", loss="mse", metrics=["mae"])

    callbacks = []
    if early_stopping:
        callbacks.append(tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True))

    history = model.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.1, callbacks=callbacks, verbose=verbose)
    
    return model, history

#Get Root Mean Squared Error (RMSE) for a given set of features
def get_rmse(feature_cols, train_df, test_df, verbose=0):
        x_train, y_train, x_test, y_test = prepare_dataset(train_df, test_df, feature_cols=feature_cols)
        tf.keras.utils.set_random_seed(2003)
        model, _ = train_model(x_train, y_train, early_stopping=True, verbose=verbose)
        y_pred = model.predict(x_test, verbose=0)
        return np.sqrt(mean_squared_error(y_test, y_pred))

#Plot training and validation metrics over epochs
def plot_train_val(history, metric="loss", title=None):
    metric = metric.lower()

    if metric == "rmse":
        if "loss" not in history.history or "val_loss" not in history.history:
            print("'loss' or 'val_loss' not found in history.")
            return
        train_values = np.sqrt(np.array(history.history["loss"]))
        val_values = np.sqrt(np.array(history.history["val_loss"]))
        label_metric = "RMSE"
        y_label = "RMSE"
    else:
        train_key = metric
        val_key = f"val_{metric}"
        if train_key not in history.history or val_key not in history.history:
            print(f"'{train_key}' or '{val_key}' not found in history.")
            return
        train_values = history.history[train_key]
        val_values = history.history[val_key]
        label_metric = metric.upper()
        y_label = metric.upper()

    plt.figure(figsize=(8, 5))
    plt.plot(train_values, label=f"Train {label_metric}")
    plt.plot(val_values, label=f"Validation {label_metric}")
    plt.xlabel("Epochs")
    plt.ylabel(y_label)
    plt.title(title or f"Train vs Validation {label_metric}")
    plt.legend()
    plt.grid(True)
    plt.show()

#Plot predicted vs true values for each target variable
def plot_pred_vs_true(y_test, y_pred, target_cols=TARGET_COLS):
    fig, axes = plt.subplots(1, len(target_cols), figsize=(8 * len(target_cols), 8))
    if len(target_cols) == 1:
        axes = [axes]

    for i, ax in enumerate(axes):
        ax.scatter(y_test[:, i], y_pred[:, i], s=10, alpha=0.5)
        mn = min(np.min(y_test[:, i]), np.min(y_pred[:, i]))
        mx = max(np.max(y_test[:, i]), np.max(y_pred[:, i]))
        ax.plot([mn, mx], [mn, mx], "r--", linewidth=1)
        ax.set_title(target_cols[i])
        ax.set_xlabel("True")
        ax.set_ylabel("Predicted")
        ax.grid(True)

    plt.tight_layout()
    plt.show()

def plot_rmse(rmse_no_es, rmse_es):
    labels = ["No Early Stopping", "Early Stopping"]
    values = [rmse_no_es, rmse_es]
    plt.figure(figsize=(6, 4))
    bars = plt.bar(labels, values)

    for bar, val in zip(bars, values):
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f"{val:.4f}", ha='center', va='bottom')
    plt.ylabel("RMSE")
    plt.title("RMSE Comparison")
    plt.grid(axis='y', alpha=0.3)
    plt.show()

def plot_elimination_history(history_df):
    df = history_df.copy().reset_index(drop=True)

    plt.figure(figsize=(9, 4.5))
    plt.plot(df.index, df["rmse"], marker="o")
    best_idx = df["rmse"].idxmin()

    plt.scatter(best_idx, df.loc[best_idx, "rmse"], s=90, zorder=3, label="Best")
    plt.xticks(df.index, df["removed"], rotation=45, ha="right")
    plt.xlabel("Feature removed at this step")
    plt.ylabel("RMSE")
    plt.title("Task 6: RMSE During Sequential Feature Elimination")
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# 1) Split the dataset so that the whole month of April is kept for testing 
#      (Note that the dataset contains only times between Feb and Apr 2013)
def sub_task_1(e2, e3, p6, omni, rbsp):
    e2 = e2.copy()
    e3 = e3.copy()
    p6 = p6.copy()
    omni = omni.copy()
    rbsp = rbsp.copy()

    dfs = [e2, e3, p6, omni, rbsp]

    for df in dfs:
        df["Time"] = pd.to_datetime(df["Time"], dayfirst=True)
    
    combined_df = (pd.concat([df.set_index("Time") for df in dfs], axis=1, join="inner")
                     .reset_index()
                     .sort_values("Time")
                     .reset_index(drop=True)
                     )
    
    is_april = (combined_df["Time"].dt.month == 4 ) & (combined_df["Time"].dt.year == 2013)
    train_df = combined_df.loc[~is_april].reset_index(drop=True)
    test_df = combined_df.loc[is_april].reset_index(drop=True)

    print(f"train rows: {len(train_df)}")
    print(f"test rows: {len(test_df)}")

    return train_df, test_df

train_df_unprocessed, test_df_unprocessed = sub_task_1(e2, e3, p6, omni, rbsp)
```

```python
# 2) Clean the dataset (remove any NaN)
def sub_task_2(train_df, test_df):
    train_len = len(train_df)
    test_len = len(test_df)
    train_df = train_df.dropna()
    test_df = test_df.dropna()
    print(f"After cleaning, train rows: {len(train_df)}, {train_len - len(train_df)} removed")
    print(f"After cleaning, test rows: {len(test_df)}, {test_len - len(test_df)} removed")

    return train_df, test_df

train_df, test_df = sub_task_2(train_df_unprocessed, test_df_unprocessed)
```

```python
# 3) Train a neural network model
def sub_task_3(x_train, y_train, x_test, y_test):

    model, history = train_model(x_train, y_train, early_stopping=False, verbose=0)

    test_loss, test_mae = model.evaluate(x_test, y_test, verbose=0)
    print(f"Test Loss: {test_loss}, Test MAE: {test_mae}")

    y_pred = model.predict(x_test, verbose=0)
    return model, history, y_pred

x_train, y_train, x_test, y_test = prepare_dataset(train_df, test_df)
model, history, y_pred = sub_task_3(x_train, y_train, x_test, y_test)
plot_train_val(history, metric="loss", title="Training + Validation Loss")
plot_train_val(history, metric="mae", title="Training + Validation MAE")
plot_pred_vs_true(y_test, y_pred)
```

```python
# 4) Compare models with and without early stopping (use RMSE as score)
def sub_task_4(x_train, y_train, x_test, y_test):
    model, history = train_model(x_train, y_train, early_stopping=False, verbose=0)
    model_early_stopping, history_early_stopping = train_model(x_train, y_train, early_stopping=True, verbose=0)

    y_pred = model.predict(x_test, verbose=0)
    y_pred_early_stopping = model_early_stopping.predict(x_test, verbose=0)

    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    rmse_early_stopping = np.sqrt(mean_squared_error(y_test, y_pred_early_stopping))

    print(f"RMSE without early stopping: {rmse}")
    print(f"RMSE with early stopping: {rmse_early_stopping}")

    return rmse, rmse_early_stopping, history, history_early_stopping

rmse, rmse_early_stopping, history_no_es, history_es = sub_task_4(x_train, y_train, x_test, y_test)

plot_train_val(history_no_es, metric="loss", title="No Early Stopping: Training + Validation Loss")
plot_train_val(history_es, metric="loss", title="Early Stopping: Training + Validation Loss")
plot_train_val(history_no_es, metric="rmse", title="No Early Stopping: Training + Validation RMSE")
plot_train_val(history_es, metric="rmse", title="Early Stopping: Training + Validation RMSE")
```

```python
# 5) Rank the features in incrasing order of importance (e.g., using a leave one feature out strategy)
def sub_task_5(train_df, test_df, target_cols=TARGET_COLS):
    feature_cols = [col for col in train_df.columns if col not in ["Time", *target_cols]]

    baseline_rmse = get_rmse(feature_cols, train_df, test_df, verbose=0) 

    rows = []
    for feature in feature_cols:
        reduced_cols = [col for col in feature_cols if col != feature]
        rmse = get_rmse(reduced_cols, train_df, test_df, verbose=0)
        importance = rmse - baseline_rmse
        rows.append({"feature": feature, "rmse": rmse, "importance": importance})

    ranked_df = pd.DataFrame(rows).sort_values("importance", ascending=True).reset_index(drop=True)

    print(f"Baseline RMSE with all features: {baseline_rmse}")
    print(ranked_df[["feature", "importance"]])

    return ranked_df

ranked_df = sub_task_5(train_df, test_df)
```

```python
# 6) Retrain the NN by eliminating, one at the time, the least important feature and determine what is the best set of features
def sub_task_6(train_df, test_df, ranked_df, target_cols=TARGET_COLS):
    all_features = [col for col in train_df.columns if col not in ["Time", *target_cols]]
    remove_order = ranked_df["feature"].tolist()

    current_features = all_features.copy()
    history = []

    baseline_rmse = get_rmse(current_features, train_df, test_df, verbose=0)
    best_rmse = baseline_rmse
    best_features = current_features.copy()
    history.append({"removed": "None", "n_features": len(current_features), "rmse": baseline_rmse})

    for feature in remove_order:
        if feature in current_features and len(current_features) > 1:
            current_features.remove(feature)
            rmse = get_rmse(current_features, train_df, test_df, verbose=0)
            history.append({"removed": feature, "n_features": len(current_features), "rmse": rmse})

            if rmse < best_rmse:
                best_rmse = rmse
                best_features = current_features.copy()
    
    history_df = pd.DataFrame(history)
    print(f"Best RMSE: {best_rmse} with features: {best_features}")
    print(history_df)
    
    return best_features, history_df

best_features_set, history_df = sub_task_6(train_df, test_df, ranked_df)
```

```python
# 7) Finally, compare the RMSE of the two models (with and without feature selection) on the test set
def sub_task_7(best_features_set, train_df, test_df, target_cols=TARGET_COLS):
    all_features = [col for col in train_df.columns if col not in ["Time", *target_cols]]

    rsme_all = get_rmse(all_features, train_df, test_df)
    rsme_selected = get_rmse(best_features_set, train_df, test_df)

    difference = rsme_all - rsme_selected

    print(f"RMSE with all features: {rsme_all}")
    print(f"RMSE with selected features: {rsme_selected}")
    print(f"Difference in RMSE: {difference}")

sub_task_7(best_features_set, train_df, test_df)
```
